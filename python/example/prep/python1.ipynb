```python
t = np.arange(100) /10 -1

def makeR(t, absorbtion=0.8, rate=0.5, noise=0.005, source=None):
    if source is None:
        source = np.zeros_like(t)
        source[np.where(t<=0)[0][-1]] = 0.8
    start = 0
    R = np.zeros_like(t)
    S = 0
    SS = 0
    for i in range(len(t)-1):
        dt = t[i+1]-t[i]
        S += source[i]
        SS = S * dt * rate
        S -= S * dt * rate
        R[i+1] = R[i] + SS * absorbtion - R[i] * absorbtion * dt + noise * np.random.randn() * np.sqrt(dt)
    return R

def makeA(t, start=1, start_noise=0.1, alpha=0.01, a_noise=0.005, beta=0.002, noise=0.01, source=None):
    if source is None:
        source = np.zeros_like(t)
        source[np.where(t<=0)[0][-1]] = 0.8
    start = start + start_noise * np.random.randn()
    A = np.zeros_like(t)
    A[0] = start
    dx = 0
    for i in range(len(t)-1):
        dt = t[i+1]-t[i]
        start -= beta * dt * start * (1 + a_noise * np.random.randn())
        start = max((start - source[i] * (1+ alpha * np.random.randn()), 0.))
        A[i+1] = A[i] - dt * A[i] * ((A[i]-start)) + \
                 noise * np.random.randn() * np.sqrt(dt) + dt*0.01
    return A

def makeB(t, max_=1, min_=-1, noise=0.02, dist=1, acc=5, source=None):
    if source is None:
        source = np.zeros_like(t)
        source[np.where(t<=0)[0][-1]] = 0.8
    def randlevel(max_,min_):
        return (max_-min_)*np.random.rand()+min_
    B = np.zeros_like(t)
    B[0] = randlevel(max_,min_)
    next_t = np.random.exponential(dist)
    for i in range(len(t)-1):
        dt = t[i+1]-t[i]
        next_t -= dt * (1 + acc * source[i])
        if next_t <= 0:
            B[i+1] = randlevel(max_,min_)
            next_t = np.random.exponential(dist)
        else:
            B[i+1] = B[i]
    return B + noise * np.random.randn(len(t))

def makeC(t, source=None):
    # before /after step
    if source is None:
        source = np.zeros_like(t)
        source[np.where(t<=0)[0][-1]] = 0.8
    delay = 2 + np.random.rand()*2
    dt = t[1] - t[0]
    l = int(delay/dt) + 20
    b1 = np.ones(l)
    b1[:10] = 0
    b1[-10:] = 0
    C = np.convolve(source,b1)[10:len(t)+10]
    ll = int(2/dt)+1
    x = np.linspace(-1,1,ll)
    g = np.exp(-x**2*25)
    C = np.convolve(C,g)[ll//3:len(t)+ll//3]
    C /= np.max(C)
    C += np.convolve(np.random.randn(len(t)-4)*0.02,[0.5,1,1,1,0.5])
    C[np.where(C<0)] = 0.
    C[np.where(C>1)] = 1.
    return C
```

```python
def make_set_2():
    rn = np.random.rand
    form = np.random.randint(3)
    if form==0:
        t = np.linspace(0,10,1001)
        source = np.zeros_like(t)
        source[100] = 0.80
    elif form == 1:
        t = np.linspace(0,20,1001)
        source = np.zeros_like(t)
        source[100] = 0.35 + 0.1*rn()
        source[300] = 0.35 + 0.1*rn()
    else:
        t = np.linspace(0,50,1001)
        source = np.zeros_like(t)
        source[100+np.random.randint(25)] = 0.15 + 0.05*rn()
        source[325+np.random.randint(25)] = 0.15 + 0.05*rn()
        source[550+np.random.randint(25)] = 0.15 + 0.05*rn()
        source[775+np.random.randint(25)] = 0.15 + 0.05*rn()
        
    
    R = makeR(t, absorbtion=0.4+0.8*rn(), rate=0.1+0.4*rn(), noise=0.005, source=source)
    A = makeA(t, start=1, start_noise=0.2, 
              alpha=0.005 + 0.01*rn(), a_noise=0.005, 
              beta=0.002 + 0.006*rn(), noise=0.01, source=R)
    B = makeB(t, max_=0, min_=10, noise=0.1, dist=0.7+rn(), acc=3+4*rn(), source=R)
    C = makeC(t, source=source)
    return np.vstack([t,source,R,A,C,B]).T 
```
